```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the News dataset
news_articles=pd.read_csv('Sentiment_dataset.csv')
news_articles.head()
```

## Preprocess the news articles

```python
news_articles['text'] = news_articles['text'].fillna('')
```

## Create a TF-IDF matrix

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(news_articles['text'])
```

## Calculate the cosine similarity matrix

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

## Define a function to get the top n most similar articles to a given article

```python
def get_top_n_similar(title, cosine_sim=cosine_sim, n=10):
    # Get the index of the article that matches the title
    idx = news_articles[news_articles['news_title'] == title].index
    if not idx.empty:
        idx = idx[0]
        # Get the pairwsie similarity scores of all articles with that article
        sim_scores = list(enumerate(cosine_sim[idx]))

        # Sort the articles based on the similarity scores
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Get the scores of the 10 most similar articles
        sim_scores = sim_scores[1:n+1]

        # Get the article indices
        article_indices = [i[0] for i in sim_scores]

        # Return the top n most similar articles
        return news_articles.iloc[article_indices]
    else:
        return pd.DataFrame()
```

## Filter the news articles based on sentiment

```python
positive_articles = news_articles[news_articles['sentiment'] == 1]
```

## Recommend the top positive articles to the user

```python
user_article = 'This is a positive news article.'## Here user can provide their article.
recommended_articles = get_top_n_similar(user_article, n=5)

if not recommended_articles.empty:
    print('Recommended Positive Articles:')
    print(recommended_articles[['news_title', 'text']])
else:
    print('No similar articles found.')
```
